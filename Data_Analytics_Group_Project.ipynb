# 🆕 New Methods & Parameters Used

1. **set_style()**    - changes the background style of the plot
2. **rcParams**       - to set default settings in matplotlib
3. **load_dataset()** - to loads a sample dataset  




```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Set style for all plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Load dataset 1: Titanic
titanic = sns.load_dataset("titanic")

# Load dataset 2: Tips
tips = sns.load_dataset("tips")

# Display top rows of each dataset
print("Titanic Dataset:")
display(titanic.head())

print("\nTips Dataset:")
display(tips.head())
```

# 1️⃣ Distribution Plots

These plots help explore how data is distributed.

We used Seaborn to visualize the distribution of numerical variables using:
- **Displot**: Combines histogram with optional KDE
- **Jointplot**: Shows scatter + distributions
- **Pairplot**: All variable pair comparisons with hue
- **Rugplot**: Marks individual data points
- **KDEPlot**: Smooth distribution curves

These plots help us understand how values are spread and detect outliers or patterns.

[Distribution Plots Documentaion - Seaborn](https://seaborn.pydata.org/tutorial/distributions.html)

# 📌 Displot📈 – Histogram with KDE

**Displot** is used to show how a numeric column is distributed. It helps us understand the shape and spread of data.

This plot type is useful to :


*   Spot patterns
*   See skewness or gaps in data
*   Detect outliers



---


**What this Shows 👇**

* This shows the age distribution of Titanic passengers:

* The bars show how many passengers are in each age group.

* The KDE (Kernel Density Estimate) curve (smooth line) shows the overall shape of the data.







```python
# Distribution of Age in Titanic dataset
#KDE curve is not necessary for this
sns.displot(data=titanic, x="age", kde=True, color="skyblue")
plt.title("Distribution of Passenger Ages - Titanic")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

# 📌 Jointplot📉 – Scatter + Histogram + KDE

**Jointplot** is used to visualize the relationship between two numeric variables. It combines a scatter plot with histogram or KDE plots.

This plot type is useful to :



*   Check if two values are related
*   Spot clusters or trends in data
*   Understand both individual and joint distributions



---

**What this Shows 👇**




*   checking the relationship between total bill and tip in a restaurant
*   The scatter plot in the middle shows how tip changes as the bill increases
*   The side histograms show the distribution of total bills and tips separately.








```python
# Relation between total bill and tip
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter", color="orange")
plt.suptitle("Joint Distribution: Total Bill vs Tip", y=1.02)   # y - to set the vertical position of the title
plt.show()
```

# 📌 Pairplot📈 - Pairwise comparisons of numerical columns


**Pairplot** is used to visually compare all numeric columns in a dataset with each other. It creates a matrix of scatter plots for every possible pair of numerical features.

This plot type is useful to :



*   Spot relationships or trends between variables
*   See how data is grouped
*   Detect clusters or outliers



---

**What this Shows 👇**



*   comparing multiple numeric columns (like total_bill, tip, size) from the Tips dataset
*   Each plot shows how two variables relate to each other
*   The diagonal shows histograms for each variable.
*   Colors (from hue) show if the customer is male or female.







```python
# Pairwise plots in Tips dataset
sns.pairplot(data=tips, hue="sex")   # hue - Helps to compare groups(male, female)
plt.suptitle("Pairplot of Tips Dataset", y=1.02)
plt.show()
```

# 📌 Rugplot📉 – Data points on an axis

**Rugplot** is used to show individual data points along a single axis (usually the X-axis). Each data point is shown as a small vertical line.


This plot type useful to:



*   See the exact values of data points
*   Check where data is dense (crowded) or sparse



---

**What this Shows 👇**



*   This shows the ages of Titanic passengers
*   Each tiny green line represents one person's age
*   If lines are close together, many people have similar ages






```python
# Rugplot for Age
sns.rugplot(data=titanic, x="age", height=0.1, color="green")
plt.title("Rugplot of Passenger Ages - Titanic")
plt.show()
```

# 📌 KDE plot📈 – Smoothed density plot

**KDE plot** stands for Kernel Density Estimate plot. It shows a smooth curve that represents the distribution of a numeric column.

This plot type useful to:


*   Understand the shape of the data
*   Spot peaks (common values) and spread (how wide values are)
*   Compare with histogram but in a smooth format



---

**What this Shows 👇**



*   This shows the distribution of total bill amounts in the Tips dataset
*   The smooth curve shows where most total bills are
*   For example, if the curve is highest near 15–20, many bills are in that range







```python
# KDE for Total Bill
sns.kdeplot(data=tips, x="total_bill", shade=True, color="purple", fill=True)
plt.title("KDE of Total Bill - Tips Dataset")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

# 2️⃣ Categorical Data Plots

Seaborn offers great support for visualizing categorical variables:
- **Boxplot**: Shows median, quartiles, and outliers.
- **Stripplot**: Displays individual data points.
- **Barplot**: Aggregated means (or other stats) with confidence intervals.
- **Countplot**: Counts the frequency of categories.

These plots reveal category-based distributions, differences, and trends.

[Categorical Plots Documentation - Seaborn](https://seaborn.pydata.org/tutorial/categorical.html)

# 📌 Boxplot📈 - Distribution with quartiles & outliers

**Boxplot** shows the distribution of a numeric column. It also highlights outliers.

This plot type useful to:


*   Understand minimum, maximum, median values
*   See how data is spread
*   Spot outliers quickly (dots outside the box)



---

**What this Shows 👇**



*   We are *comparing the total bill amounts across different days* in the Tips dataset
*   The line inside the box shows the median
*   Dots outside the box are outliers








```python
# Boxplot of total bill by day
sns.boxplot(data=tips, x="day", y="total_bill", palette="Set2")   # palette = "Set2" --> Set a color theme for the boxes
plt.title("Boxplot: Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

# 📌 Stripplot📉 - Individual data points by category

**Stripplot** is used to show all individual data points for each category. It plots small dots for each value, giving a closer look at raw data.

This plot type useful to:



*   See the exact values and how they are spread
*   Spot clusters or patterns in small datasets



---

**What this Shows 👇**



*   We are showing total bill amounts for each day
*   A boxplot to show the overall distribution
*   A stripplot on top to show each individual bill




```python
# Stripplot on top of boxplot
sns.boxplot(data=tips, x="day", y="total_bill", color="lightgray")
sns.stripplot(data=tips, x="day", y="total_bill", color="red", jitter=True)   # jitter=True: Spreads dots slightly so they don’t overlap.
plt.title("Stripplot over Boxplot: Total Bill by Day")
plt.show()
```

# 📌 Barplot📈 - Aggregated values with error bars

**Barplot** shows summary statistics (like average, sum, count) for different categories. By default, it shows the mean value with error bars (variation).

This plot type useful to:


*   Compare average values across categories
*   Visualize differences between groups
*   Understand variation using error bars



---
**What this Shows 👇**



*   Shows the average tip given by males and females in the Tips dataset
*   The height of the bar shows the mean tip for each gender
*   The error bar (black line on top) shows how much the values can vary






```python
# Average tip by gender
sns.barplot(data=tips, x="sex", y="tip", estimator="mean", palette="pastel")
plt.title("Barplot: Average Tip by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Tip")
plt.show()
```

# 📌 Countplot📉 - Frequency counts

**Countplot** shows how many times each category appears in the data. It counts the number of records for each category and shows it as bars.

This plot type useful to:



*   See the frequency of each group
*   Understand how data is distributed by category
*   Compare sizes of different categories easily



---

**What this Shows 👇**



*   This shows the number of Titanic passengers in each class (First, Second, Third)
*   Each bar shows how many people traveled in each class.





```python
# Number of passengers in each class
sns.countplot(data=titanic, x="class", palette="viridis")
plt.title("Count of Passengers by Class - Titanic")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

# 3️⃣ Matrix Plots

Matrix plots help visualize large data relationships in a compact table format.

- **Correlation Matrix**: Shows how strongly variables are linearly related.
- **Pivot Table + Heatmap**: Aggregates values (like averages) across two categorical variables.

Heatmaps make it easy to spot high/low correlations and trends in tabular data.

# 📌 Heatmap📈 - Correlation Matrix

**Heatmap** is a colored table used to show relationships between numeric columns. When used with a correlation matrix, it shows how strongly two variables are related.

This plot type is useful to:



*   Understand which columns are connected
*   Spot positive or negative correlations



---


**What this Shows 👇**



*   This calculates the correlation between all numeric columns in the Titanic dataset
*   A value close to +1 means strong positive correlation (they increase together).
*   A value close to –1 means strong negative correlation (one increases, other decreases)
*   A value close to 0 means no strong relationship.



---



The heatmap displays these values with colors:

Red 🔴 = Positive correlation

Blue 🔵 = Negative correlation

White ⚪ = No correlation


[Heat Maps Documentation - Seaborn](https://seaborn.pydata.org/generated/seaborn.heatmap.html)


```python
# Compute correlation matrix for numerical columns  (Taitanic dataset)
titanic_corr = titanic.corr(numeric_only=True)

# Create heatmap
sns.heatmap(titanic_corr, annot=True, cmap="coolwarm", linewidths=0.5)  # cmap - color theme , annot - Shows the numeric correlation values on the plot.
plt.title("Correlation Matrix - Titanic Dataset")
plt.show()
```

# 📌 Heatmap📉 - based on Pivot Table

**Heatmap** shows a summary table (pivot table) where the values are aggregated (like mean, sum) across two categories. The heatmap then colors the table to make patterns easier to spot.

This plot type useful to:


*   Compare averages or totals between two categories
*   Quickly understand where values are high or low
*   Make pivot tables easier to read with color



---

**What this Shows 👇**



*   This shows tips dataset to see the average total bill for (day of the week and time)
*   Darker colors = Higher average bills
*   Lighter colors = Lower average bills





```python
# Create a pivot table   (tips dataset) - Average total bill by day and time
pivot_tips = tips.pivot_table(values="total_bill", index="day", columns="time", aggfunc="mean")

# Heatmap of pivot table
sns.heatmap(pivot_tips, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Average Total Bill by Day and Time - Tips Dataset")
plt.show()
```

# 4️⃣ Grid-Based Plots

Grid plots help break down visualizations by rows, columns, and hue for deeper analysis.

- **FacetGrid**: Create subplots across rows and columns based on categorical values.
- **PairGrid**: Flexible version of pairplot, lets us customize each subplot.
- **JointGrid**: Layout for joint distributions with individual axis plots.

These grids reveal patterns across subsets of data by categories like gender, smoker, time etc.

[Grid Based Plot Documentation - Seaborn](https://seaborn.pydata.org/tutorial/axis_grids.html)

# 📌 FacetGrid📈 – Multiple plots based on a category

**FacetGrid** lets you create multiple small plots (subplots), each based on combinations of categories. It helps you compare patterns across groups side by side.

This plot type useful to:


*   Analyze how a variable behaves across different categories
*   Break down data visually into smaller segments
*   Spot trends or differences between groups easily



---

**What this Shows 👇**

In this plot, we’re showing the distribution of total bills in the Tips dataset:

* Separated by gender (columns)

* smoker status (rows)



```python
# Plot total bill by gender and smoker status
g = sns.FacetGrid(tips, col="sex", row="smoker", margin_titles=True)
g.map(sns.histplot, "total_bill", kde=True, color="skyblue")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("FacetGrid: Total Bill by Gender and Smoker")
plt.show()
```

# 📌 PairGrid📉 – Similar to pairplot but more flexible

**PairGrid** is like pairplot, but it gives you more control over how each part of the grid is displayed. You can customize the diagonal, upper, and lower plots separately.

This plot type useful to:


*   Create custom pairwise comparisons
*   Show relationships between multiple numeric columns grouped by a category



---
**What this Shows 👇**



*   Use the Tips dataset and compare numeric columns (like total_bill, tip, size)
*   Diagonal shows individual distributions (histograms)
*   Off-diagonal shows relationships between columns (scatter plots)






```python
# Custom pairwise plots for numerical features in Tips
g = sns.PairGrid(tips, hue="sex")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.suptitle("PairGrid: Tips Dataset by Gender", y=1.02)
plt.show()
```

## 📌 JointGrid📈 – Custom jointplot layout

**JointGrid** is a more customizable version of jointplot.


This plot type useful to:


*   You want to mix different plot types
*   You need more control over layout and styling than jointplot provides



---

**What this Shows 👇**


*   Explore the relationship between total_bill (x-axis) & tip (y-axis)
*   Center: Scatter plot (relationship between total bill and tip)
*   Top & Right: Histograms of total bill and tip values






```python
# Custom joint plot using JointGrid
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.scatterplot, sns.histplot)
g.fig.suptitle("JointGrid: Total Bill vs Tip", y=1.02)
plt.show()
```

# 5️⃣ Regression Plots

Regression plots in Seaborn visualize linear relationships between two continuous variables:
- **lmplot** draws regression lines with optional confidence intervals.
- We used `hue`, `col`, and `row` to add extra categorical dimensions.
- The plots help us understand trends like: Do men tip more? Does time affect the relationship?

[Regression Plots Documentation - Seaborn](https://seaborn.pydata.org/tutorial/regression.html)


# 📌 lmplot📈 – Basic linear regression (Total bill vs Tip)

lmplot is used to draw a straight line that best fits the data   (linear regression).

This plot type useful to:


*   Understand the relationship between two numeric variables
*   See if one variable increases or decreases as the other increases



---

**What this Shows 👇**

* We explore the relationship between total_bill & tips
* Scatter plot of all the data points
* Regression line showing the trend (best-fit line)



```python
sns.lmplot(data=tips, x="total_bill", y="tip", height=5, aspect=1.2)    # height -  Controls the height of the plot  ,  aspect - Controls the width relative to height
# By default it adds scatterplot and best fit line
plt.title("Linear Regression: Total Bill vs Tip")
plt.show()
```

# 📌 lmplot with hue & markers📉 – Regression lines by Gender

**What** **this** **shows** 👇



*   Explore the relationship between total_bill (x-axis) and tip (y-axis).
*   Color (hue): Gender (Male/Female) — shown using different colors from "Set1".

*   Markers: Different shapes ("o", "s") to make categories clearer.
*   Regression lines: Separate for each gender.







```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex",markers=["o", "s"], palette="Set1", height=6)
plt.title("Regression by Gender: Total Bill vs Tip")
plt.show()
```

# 📌 lmplot with col & row 📈 – Multiple Regression Plots (Time & Gender)

**What** **this** **shows** 👇


*   Relationship between total_bill (x-axis) and tip (y-axis).
*   Columns (col): Split by Time (Lunch / Dinner) horizontally.

*   Rows (row): Split by Gender (Male / Female) vertically.
*   4 Separate plots: Each combination of Gender & Time has its own regression line.







```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time",row="sex", height=4, aspect=1)
plt.suptitle("Regression Plots by Time and Gender", y=1.05)
plt.show()
```

# 7️⃣ Plot Customization

Seaborn offers several tools to improve the look and readability of plots:
- **set_style()**: Background grid styling
- **set_context()**: Font scaling for various uses (paper, talk, poster)
- **palette**: Predefined or custom color themes
- **despine()**: Removes top and right axis lines
- **height/aspect**: Controls size and shape of plots

Good customization makes insights clearer and visuals more engaging.

# a. set_style() – Change plot background style

```python
# Available styles: "whitegrid", "darkgrid", "white", "dark", "ticks"
sns.set_style("darkgrid")

sns.boxplot(data=tips, x="day", y="total_bill", palette="Set3")
plt.title("Styled Boxplot with Darkgrid")
plt.show()
```

# b. set_context() – Scale font size for context

```python
# Contexts: "paper", "notebook", "talk", "poster"
sns.set_context("talk")

sns.histplot(data=tips, x="total_bill", kde=True, color="teal")
plt.title("Talk Context: Histogram of Total Bill")
plt.show()
```

# c. palette – Change color palette

```python
sns.set_palette("coolwarm")

sns.countplot(data=titanic, x="class")
plt.title("Count of Passengers by Class with Coolwarm Palette")
plt.show()
```

# d. despine() – Remove plot spines

```python
sns.barplot(data=tips, x="sex", y="tip", palette="pastel")
sns.despine()
plt.title("Barplot with Despined Axes")
plt.show()
```

# e. height & aspect – Control figure size in lmplot, FacetGrid, etc.

```python
sns.lmplot(data=tips, x="total_bill", y="tip", height=5, aspect=1.5)
plt.title("Custom Size lmplot (5x1.5)")
plt.show()
```
